```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt             
%matplotlib inline
import seaborn as sns
```

```python
plt.rcParams['figure.figsize']
```

```python
url = 'https://vincentarelbundock.github.io/Rdatasets/csv/itsadug/eeg.csv'

df = pd.read_csv(url,index_col=0) # , usecols=['time','value'], index_col=['time'])
display(df.head())

fig, ax = plt.subplots()
sns.lineplot(x=df.Time, y=df.Ampl, ax=ax)
ax.set_title(url.split('/')[-1].split('.csv')[0])
plt.show()
```

```python
plt.style.use([ 'grayscale' , '/content/tsp.mplstyle' ])

url = 'https://vincentarelbundock.github.io/Rdatasets/csv/itsadug/eeg.csv'

df = pd.read_csv(url,index_col=0) # , usecols=['time','value'], index_col=['time'])
display(df.head())

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(16,5))
sns.lineplot(x=df.Time, y=df.Ampl, ax=ax1)
ax1.set_title(url.split('/')[-1].split('.csv')[0])

sns.lineplot(x=df.Time, y=-df.Ampl, ax=ax2)
ax2.set_title('-' + url.split('/')[-1].split('.csv')[0])
plt.show()
```
